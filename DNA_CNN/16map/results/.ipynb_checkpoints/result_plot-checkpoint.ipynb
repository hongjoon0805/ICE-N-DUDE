```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PI_type_arr = ['A', 'B', 'C']
BW_1st_arr = [np.ones(20)*0.78, np.ones(20)*0.82, np.ones(20)*0.81]
for i in range(3):
    PI_type = PI_type_arr[i]
    k_arr = np.arange(10, 201, 10)
    BW_1st = BW_1st_arr[i]
    schemes = ['N-DUDE', 'ICE-N-DUDE']
    file = PI_type + '_DNA_results.mat'
    
    N_DUDE = sio.loadmat(file)[schemes[0]]
    ICE_N_DUDE = sio.loadmat(file)[schemes[1]]
                
    plt.figure()
    
    plt.plot(k_arr, BW_1st, linestyle = '-', label = 'BW_1st', marker = 'o')
    plt.plot(k_arr, N_DUDE[:,0], linestyle = '-', label = 'N-DUDE', marker = '*')
    plt.plot(k_arr, ICE_N_DUDE[:,0], linestyle = '-', label = 'ICE-N-DUDE', marker = '.')
    
    plt.xlim(0,200)
    plt.xticks([20,40,60,80,100,120,140,160,180,200])
    
    plt.ylim(0.5,0.9)
    plt.yticks([0.5,0.6,0.7,0.8,0.9], fontsize = 13)

    plt.xlabel("Window Size k", fontsize = 20)
    plt.ylabel("Error rate", fontsize = 20)

    plt.legend(loc = 'upper right', frameon = True)
    
    plt.title('PI type '+PI_type, fontsize = 20)
```
