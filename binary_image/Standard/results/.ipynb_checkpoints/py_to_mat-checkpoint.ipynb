```python
import numpy as np
import sys
import scipy.io as sio
```

```python
def py_to_mat(nb_x_classes, PI_type, assumed_delta):
    
    results_dict = {}
    process_name_arr = ['BW.py', 'NDUDE.py', 'ICE_N_DUDE.py']
    
    results = [np.zeros(8), np.zeros(8), np.zeros(8)]
    process_name = '%d_%s_%.2f'%(nb_x_classes, PI_type, assumed_delta) + '__BW.py'
    
    with open(process_name, 'r') as f:
        data = f.read().splitlines()
        for i in range(len(data)):
            arr = data[i].split()
            order, img, error = int(arr[0]), int(arr[1]),  float(arr[2])
            results[order-1][img] = error
    
    results_dict['BW.py'] = results
    
    for order in range(3):
        arr = results[order]
        mask = arr == 0
        if np.sum(results==0) > 1:
            print('%d_%s_%.2f'%(nb_x_classes, PI_type, assumed_delta))
            print('NDUDE Sanity Check!')
            for i in range(8):
                if mask[i] == 1:
                    print('%d'%i)
    
    results = np.zeros(8)
    process_name = '%d_%s'%(nb_x_classes, PI_type) + '_NDUDE.py'
    
    with open(process_name, 'r') as f:
        data = f.read().splitlines()
        for i in range(len(data)):
            arr = data[i].split()
            img, error = int(arr[0]), float(arr[1]) 
            results[img] = error

    results_dict['NDUDE.py'] = results
    
    mask = results == 0
    if np.sum(results==0) > 1:
        print('%d_%s_%.2f'%(nb_x_classes, PI_type, assumed_delta))
        print('NDUDE Sanity Check!')
        for i in range(8):
            if mask[i] == 1:
                print('%d'%i)
        
    
    results = np.zeros(8)
    process_name = '%d_%s_%.2f'%(nb_x_classes, PI_type, assumed_delta) + '__NDUDE_finetuning.py'
    
    with open(process_name, 'r') as f:
        data = f.read().splitlines()
        for i in range(len(data)):
            arr = data[i].split()
            img, error = int(arr[0]), float(arr[1])
            results[img] = error

    results_dict['NDUDE_finetuning.py'] = results
    
    mask = results == 0
    if np.sum(results==0) > 1:
        print('%d_%s_%.2f'%(nb_x_classes, PI_type, assumed_delta))
        print('finetuning Sanity Check!')
        for i in range(8):
            if mask[i] == 1:
                print('%d'%i)
    
    file = '%d_%s_%.2f'%(nb_x_classes, PI_type, assumed_delta) + 'quantized.py'
    sio.savemat(file, results_dict)
    return results_dict
```

```python
PI_type_arr = ['20%', '30%','10']
nb_x_classes_arr = [2]
assumed_delta_arr = [0.1]

for PI_type in PI_type_arr:
    for nb_x_classes in nb_x_classes_arr:
        for assumed_delta in assumed_delta_arr:
            py_to_mat(nb_x_classes, PI_type, assumed_delta)
```
